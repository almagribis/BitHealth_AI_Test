# Case 1: Python - Basic Data Processing

1. Load CSV File

```python
import pandas as pd
```

```python
"""load csv file"""
fpath = "data/patients.csv"
df = pd.read_csv(fpath)
```

2. Tokenizes the symptoms into lists

```python
"""show current symptoms columns"""
df.head()
```

```python
print(f"symptoms column type: {type(df.iloc[0]['symptoms'])}")
```

```python
df["symptoms_tokenize"] = df["symptoms"].str.split(",")
df.head()
```

```python
print(f"symptoms column type: {type(df.iloc[0]['symptoms_tokenize'])}")
```

3. Filters patients older than 40 with more than 2 symptoms

```python
filtered_df = df[(df["age"] > 40) & (df["symptoms_tokenize"].apply(len)>2)]
filtered_df
```
