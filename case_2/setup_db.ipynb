# Setup database hospital_record for each table : patients, visits and symptoms from csv file

```python
import pandas as pd
import sqlite3
import os
```

```python
"""setup path for db and csv file"""
DATA_DIR = "data"
DB_NAME = os.path.join(DATA_DIR, "hospital_records.db")

patients_csv = os.path.join(DATA_DIR, "patients.csv")
visits_csv = os.path.join(DATA_DIR, "visits.csv")
symptoms_csv = os.path.join(DATA_DIR, "symptoms.csv")
```

```python
"""load csv file"""
patients_df = pd.read_csv(patients_csv)
visits_df = pd.read_csv(visits_csv)
symptoms_df = pd.read_csv(symptoms_csv)
```

```python
"""convert visit_date to datetome object"""
visits_df["visit_date"] = pd.to_datetime(visits_df["visit_date"], format="%Y-%m-%d")
```

```python
"""connect to sqlite"""
conn = sqlite3.connect(DB_NAME)
```

```python
"""save all table to database"""
patients_df.to_sql("patients", conn, if_exists="replace", index=False)
visits_df.to_sql("visits", conn, if_exists="replace", index=False,
                 dtype={"visit_date": "DATE"})
symptoms_df.to_sql("symptoms", conn, if_exists="replace", index=False)
```
